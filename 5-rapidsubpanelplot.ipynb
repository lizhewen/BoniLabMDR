```
import pandas as pd
# read in IQR data
%cd 'drive/My Drive/BoniLabMDR/data/simulation-set-2'
MFT_L_IQRdf = pd.read_csv('1-MFT-L.csv')
MFT_M_IQRdf = pd.read_csv('1-MFT-M.csv')
MFT_U_IQRdf = pd.read_csv('1-MFT-U.csv')

Cyc_L_IQRdf = pd.read_csv('1-Cyc-L.csv')
Cyc_M_IQRdf = pd.read_csv('1-Cyc-M.csv')
Cyc_U_IQRdf = pd.read_csv('1-Cyc-U.csv')

AdpCyc_L_IQRdf = pd.read_csv('1-AdpCyc-L.csv')
AdpCyc_M_IQRdf = pd.read_csv('1-AdpCyc-M.csv')
AdpCyc_U_IQRdf = pd.read_csv('1-AdpCyc-U.csv')
```

```
% cd '../..'

# prepare plt for 5x3 subpanels
import matplotlib.pyplot as plt
from plot_helper import xaxis_label_ticker
import matplotlib.ticker as ticker
xlocator = 5*365
ticks_x = xaxis_label_ticker()

fig, axs = plt.subplots(5, 3, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})
fig.patch.set_facecolor('white')
fig.suptitle('Evolution of Multiple-Drug-Resistant Types', y=0.92, fontweight='bold')
plt.rcParams['figure.figsize'] = [20, 12]
(ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9), \
      (ax10, ax11, ax12), (ax13, ax14, ax15) = axs

# plot trends for most-dangerous-triple (2 types)
from plotter import geno_trend_plot_most_dange_trip, \
                    geno_trend_plot_double_higher

#  MFT
geno_trend_plot_most_dange_trip(ax1, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'TYY..Y2.')
geno_trend_plot_most_dange_trip(ax4, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'KNF..Y2.')
geno_trend_plot_double_higher(ax7, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'DHA-PPQ')
geno_trend_plot_double_higher(ax10, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'AS-AQ')
geno_trend_plot_double_higher(ax13, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'AL')

#  Cycling
geno_trend_plot_most_dange_trip(ax2, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'TYY..Y2.')
geno_trend_plot_most_dange_trip(ax5, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'KNF..Y2.')
geno_trend_plot_double_higher(ax8, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'DHA-PPQ')
geno_trend_plot_double_higher(ax11, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'AS-AQ')
geno_trend_plot_double_higher(ax14, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'AL')

#  Adaptive Cycling
geno_trend_plot_most_dange_trip(ax3, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'TYY..Y2.')
geno_trend_plot_most_dange_trip(ax6, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'KNF..Y2.')
geno_trend_plot_double_higher(ax9, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'DHA-PPQ')
geno_trend_plot_double_higher(ax12, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'AS-AQ')
geno_trend_plot_double_higher(ax15, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'AL')

# Labeling Strategies
ax1.set_title('MFT')
ax2.set_title('5-Year Cycling')
ax3.set_title('Adaptive Cycling')

ax13.xaxis.set_major_locator(ticker.MultipleLocator(xlocator))
ax13.xaxis.set_major_formatter(ticks_x)
ax14.xaxis.set_major_locator(ticker.MultipleLocator(xlocator))
ax14.xaxis.set_major_formatter(ticks_x)
ax15.xaxis.set_major_locator(ticker.MultipleLocator(xlocator))
ax15.xaxis.set_major_formatter(ticks_x)

ax13.set_xlabel('Years')
ax14.set_xlabel('Years')
ax15.set_xlabel('Years')
# ax1.set_ylabel('Genotype Frequency (%)')
# ax4.set_ylabel('Genotype Frequency (%)')
# ax7.set_ylabel('Genotype Frequency (%)')
# ax10.set_ylabel('Genotype Frequency (%)')
# ax13.set_ylabel('Genotype Frequency (%)')

plt.show()
```

```
% cd '../..'

# prepare plt for 5x3 subpanels
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fontpath = ['Open_Sans/OpenSans-Regular.ttf', 'Open_Sans/OpenSans-Bold.ttf']
fm.fontManager.ttflist += fm.createFontList(fontpath)
mpl.rc('font', family='Open Sans')

from plot_helper import xaxis_label_ticker
import matplotlib.ticker as ticker
xlocator = 5*365
ticks_x = xaxis_label_ticker()

fig, axs = plt.subplots(5, 3, sharex='col', sharey='row',
                        gridspec_kw={'hspace': 0, 'wspace': 0})
fig.patch.set_facecolor('white')
fig.suptitle('Evolution of Multiple-Drug-Resistant Types', y=0.92, fontweight='bold')
plt.rcParams['figure.figsize'] = [20, 12]
(ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9), \
      (ax10, ax11, ax12), (ax13, ax14, ax15) = axs

# plot trends for most-dangerous-triple (2 types)
from plotter import geno_trend_plot_most_dange_trip, \
                    geno_trend_plot_double_higher

# Define ylim for annotations
row1ylim = 0.015
row2ylim = 0.000006
row3ylim = 0.5
row4ylim = 0.3
row5ylim = 0.3

#  MFT
geno_trend_plot_most_dange_trip(ax1, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'TYY..Y2.', row1ylim)
geno_trend_plot_most_dange_trip(ax4, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'KNF..Y2.', row2ylim)
geno_trend_plot_double_higher(ax7, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'DHA-PPQ', row3ylim)
geno_trend_plot_double_higher(ax10, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'AS-AQ', row4ylim)
geno_trend_plot_double_higher(ax13, MFT_L_IQRdf, MFT_M_IQRdf, MFT_U_IQRdf, 'AL', row5ylim)

#  Cycling
geno_trend_plot_most_dange_trip(ax2, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'TYY..Y2.', row1ylim)
geno_trend_plot_most_dange_trip(ax5, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'KNF..Y2.', row2ylim)
geno_trend_plot_double_higher(ax8, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'DHA-PPQ', row3ylim)
geno_trend_plot_double_higher(ax11, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'AS-AQ', row4ylim)
geno_trend_plot_double_higher(ax14, Cyc_L_IQRdf, Cyc_M_IQRdf, Cyc_U_IQRdf, 'AL', row5ylim)

#  Adaptive Cycling
geno_trend_plot_most_dange_trip(ax3, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'TYY..Y2.', row1ylim)
geno_trend_plot_most_dange_trip(ax6, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'KNF..Y2.', row2ylim)
geno_trend_plot_double_higher(ax9, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'DHA-PPQ', row3ylim)
geno_trend_plot_double_higher(ax12, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'AS-AQ', row4ylim)
geno_trend_plot_double_higher(ax15, AdpCyc_L_IQRdf, AdpCyc_M_IQRdf, AdpCyc_U_IQRdf, 'AL', row5ylim)

# Labeling Strategies
ax1.set_title('MFT')
ax2.set_title('5-Year Cycling')
ax3.set_title('Adaptive Cycling')

ax13.xaxis.set_major_locator(ticker.MultipleLocator(xlocator))
ax13.xaxis.set_major_formatter(ticks_x)
ax14.xaxis.set_major_locator(ticker.MultipleLocator(xlocator))
ax14.xaxis.set_major_formatter(ticks_x)
ax15.xaxis.set_major_locator(ticker.MultipleLocator(xlocator))
ax15.xaxis.set_major_formatter(ticks_x)

ax13.set_xlabel('Years')
ax14.set_xlabel('Years')
ax15.set_xlabel('Years')
ax1.set_ylabel('Genotype Frequency',rotation=0, labelpad=20)
ax4.set_ylabel('Genotype Frequency',rotation=0, labelpad=20)
ax7.set_ylabel('Genotype Frequency',rotation=0, labelpad=20)
ax10.set_ylabel('Genotype Frequency',rotation=0, labelpad=20)
ax13.set_ylabel('Genotype Frequency',rotation=0, labelpad=20)

plt.savefig(fname='/pic1.svg', format='svg')
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
df = pd.read_csv('monthly_data_0.txt', sep='\t') # convert tsv to pandas data frame
```

```python
# scale func to show x-axis in years
scale_x = 365
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_x))
```

```python
# blood_slide_prevalence plot
fig1 = plt.figure(1)
ax1 = fig1.add_subplot(111)
ax1.plot(df['current_time'], df['blood_slide_prev'])
ax1.xaxis.set_major_locator(ticker.MultipleLocator(1825))
ax1.xaxis.set_major_formatter(ticks_x)
#fig.show()
```

```python
# sum up total num of parasites in the silumation
tot_para = df.iloc[:,22:150].sum().sum()
```

```python
fig2 = plt.figure(2)
ax2 = fig2.add_subplot(111)
# resistance group to AS-AQ (TNY--Y**, 68-71)
ax2.plot(df['current_time'], df['TNY--Y1x']+df['TNY--Y1X']+df['TNY--Y2x']+df['TNY--Y2X'], label='AS-AQ')
# resistance group to AL (KNF--Y**, 20-23)
ax2.plot(df['current_time'], df.iloc[:,0:151].filter(regex='KNF--Y..', axis=1).sum(axis=1), label='AL')
# resistance group to DHA-PPQ (***--Y2*)
ax2.plot(df['current_time'], df.iloc[:,0:151].filter(regex='...--Y2.', axis=1).sum(axis=1), label='DHA-PPQ')

# format x-axis
ax2.xaxis.set_major_locator(ticker.MultipleLocator(1825))
ax2.xaxis.set_major_formatter(ticks_x)
# format y-axis
ax2.yaxis.set_major_formatter(ticker.PercentFormatter(xmax = tot_para))

ax2.legend()
```

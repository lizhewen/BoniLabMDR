## Code Demo 1: txt to dataframe

Config: 40-yr-simu, 50,000-popu, 0.7-treatment-cvrg, DHAPPQ-ASAQ-AL-cycling

```python
import pandas as pd

df = pd.read_csv('monthly_data_0.txt', sep='\t')
df
```

## Code Demo 2: Plot using `matplotlib`, and `ticker` function

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

plt.rcParams['figure.figsize'] = [20, 10]
fig, ax1 = plt.subplots()

scale_x = 365
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_x))

ax1.plot(df['current_time'], df['blood_slide_prev'])
ax1.xaxis.set_major_locator(ticker.MultipleLocator(5*365))
ax1.xaxis.set_major_formatter(ticks_x)
ax1.set_xlabel('years')
ax1.set_ylabel('blood slide prev')
ax1.set_title('Blood Slide Prevalence Plot')
```

## Code Demo 4: Plotting by genotypes and subplots

```python
# blood_slide_prevalence plot
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot(df['current_time'], df['blood_slide_prev'])
ax1.xaxis.set_major_locator(ticker.MultipleLocator(1825))
ax1.xaxis.set_major_formatter(ticks_x)
ax1.set_xlabel('years')
ax1.set_ylabel('blood slide prev')
ax1.set_title('Blood Slide Prev. Plot')

# parasite freq plot
ax2 = fig.add_subplot(212)
# grouped comparisons for DHA-PPQ resistance
ax2.plot(df['current_time'], df.iloc[:,0:151].filter(regex='.....C1.', axis=1).sum(axis=1), label='*C1.(wild)')
ax2.plot(df['current_time'], df.iloc[:,0:151].filter(regex='.....C2.', axis=1).sum(axis=1), label='*C2.(PPQ)')
ax2.plot(df['current_time'], df.iloc[:,0:151].filter(regex='.....Y1.', axis=1).sum(axis=1), label='*Y1.(Artim.)')
ax2.plot(df['current_time'], df.iloc[:,0:151].filter(regex='.....Y2.', axis=1).sum(axis=1), label='*Y2.(double)')

# format x-axis
ax2.xaxis.set_major_locator(ticker.MultipleLocator(1825))
ax2.xaxis.set_major_formatter(ticks_x)
# format y-axis
ax2.set_xlabel('years')
ax2.set_ylabel('parasite freq')
ax2.set_title('Parasite Freq. Plot')
ax2.legend()
```

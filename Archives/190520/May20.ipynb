```python
import matplotlib.pyplot as plt
import csv
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
# two times plot with strategy #1 (AL-SFT)
```

```python
x = []
y = []
with open('1.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x.append(float(row[0]))
        y.append(float(row[12]))
x = [i/365 for i in x]
```

```python
plt.plot(x,y)
x = []
y = []
with open('2.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter='\t')
    for row in plots:
        x.append(float(row[0]))
        y.append(float(row[12]))
x = [i/365 for i in x]
plt.plot(x,y)
plt.show()
```

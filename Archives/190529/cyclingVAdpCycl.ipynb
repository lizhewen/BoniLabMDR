```python
import matplotlib.pyplot as plt
import csv
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
# ten times of using normal cycling
```

```python
for i in range(1,11):
    x = []
    y = []
    filepath = 'cycling/%d.txt' % i
    with open(filepath,'r') as csvfile:
        plots = csv.reader(csvfile, delimiter='\t')
        for row in plots:
            x.append(float(row[0]))
            y.append(float(row[12]))
    x = [i/365 for i in x]
    plt.plot(x,y)
plt.show()
```

```python
# ten times of using adaptive cycling
```

```python
for j in range(1,11):
    x1 = []
    y1 = []
    filepath2 = 'adapCycling/%d.txt' % j
    with open(filepath2,'r') as csvfile:
        plots = csv.reader(csvfile, delimiter='\t')
        for row in plots:
            x1.append(float(row[0]))
            y1.append(float(row[12]))
    x1 = [i/365 for i in x1]
    plt.plot(x1,y1)
plt.show()
```

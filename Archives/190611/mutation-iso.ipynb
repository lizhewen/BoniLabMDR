```python
import matplotlib.pyplot as plt
import csv
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
for i in range(1):
    x = []
    y = []
    y14 = []
    y15 = []
    y76 = []
    y77 = []
    y78 = []
    #filepath = 'cycling/%d.txt' % i
    with open('monthly_data_0.txt','r') as csvfile:
        plots = csv.reader(csvfile, delimiter='\t')
        for row in plots:
            x.append(float(row[0]))
            y.append(float(row[12]))
            y14.append(float(row[36]))
            y15.append(float(row[37]))
            y76.append(float(row[98]))
            y77.append(float(row[99]))
            y78.append(float(row[100]))
    x = [i/365 for i in x]
    plt.plot(x,y)
    #plt.plot(x,y14)
    #plt.plot(x,y15)
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
plt.rcParams['figure.figsize'] = [20, 15]
```

```python
df = pd.read_csv('0.txt', sep='\t') # convert tsv to pandas data frame
df1 = pd.read_csv('1.txt', sep='\t')
df2 = pd.read_csv('2.txt', sep='\t')
```

```python
# scale func to show x-axis in years
scale_x = 12
ticks_x = ticker.FuncFormatter(lambda x, pos: '{0:g}'.format(x/scale_x))
```

```python
# blood_slide_prevalence plot
fig = plt.figure()
ax1 = fig.add_subplot(211)
ax1.plot(df['blood_slide_prev'])
ax1.xaxis.set_major_locator(ticker.MultipleLocator(12*5))
ax1.xaxis.set_major_formatter(ticks_x)
ax1.set_xlabel('years')
ax1.set_ylabel('blood slide prev')
ax1.set_title('Blood Slide Prev. Plot #0')

# sum up total num of parasites in the silumation
tot_para = df.iloc[:,22:150].sum(axis=1)
# parasite freq plot
ax2 = fig.add_subplot(212)
# grouped comparisons for DHA-PPQ resistance
ax2.plot(df.iloc[:,0:151].filter(regex='.....C1.', axis=1).sum(axis=1)/tot_para, label='wild')
ax2.plot(df.iloc[:,0:151].filter(regex='.....C2.', axis=1).sum(axis=1)/tot_para, label='PPQ-res')
ax2.plot(df.iloc[:,0:151].filter(regex='.....Y1.', axis=1).sum(axis=1)/tot_para, label='Artim.-res')
ax2.plot(df.iloc[:,0:151].filter(regex='.....Y2.', axis=1).sum(axis=1)/tot_para, label='PPQ-Art-double')
ax2.plot(df.iloc[:,0:151].filter(regex='TY......', axis=1).sum(axis=1)/tot_para, label='AQ-res')
ax2.plot(df.iloc[:,0:151].filter(regex='KN......', axis=1).sum(axis=1)/tot_para, label='Lum-res')

# format x-axis
ax2.xaxis.set_major_locator(ticker.MultipleLocator(12*5))
ax2.xaxis.set_major_formatter(ticks_x)
# format y-axis
#ax2.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1.0))
ax2.set_xlabel('years')
ax2.set_ylabel('parasite freq')
ax2.set_title('Parasite Freq. Plot #0')
ax2.legend()
```

```python
#                          DHAPPQ      ASAQ      AL       DHAPPQ     ASAQ       AL      DHAPPQ     ASAQ       AL
```
